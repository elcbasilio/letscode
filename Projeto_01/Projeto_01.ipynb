```python
import csv

def tela_principal():
    print ('+---------------------------------------------------------------------+')
    print ('|                           Controle de Estoque                       |')
    print ('|---------------------------------------------------------------------|')
    print ('|                                                                     |')
    print ('|                              Opções:                                |')
    print ('|                                                                     |')
    print ('|               1 - Inserir um item                                   |')
    print ('|               2 - Visualizar estoque                                |')
    print ('|               3 - Remover um item pelo nome                         |')
    print ('|               4 - Filtrar um item pelo nome                         |')
    print ('|               5 - Salvar o estoque atual em arquivo                 |') 
    print ('|               6 - Carregar                                          |')
    print ('|               7 - Sair                                              |')
    print ('|                                                                     |')
    print ('|---------------------------------------------------------------------|')
    opcao = int (input('Digite a opção desejada: '))
    return opcao

def item1():
    cadastro=[]
    naosair = 'S'
    while naosair.upper() == 'S':
        print ('+---------------------------------------------------------------------+')
        print ('|                     Adicionando itens ao estoque                    |')
        print ('+---------------------------------------------------------------------+')
        nome = input ('Digite o nome do item: ')
        cadastro.append(nome)
        atualizei=False
        for item in cadastro_total:
            if item[0] == nome:
                qtde = int (input ('Digite a quantidade de produtos deste item: '))
                item[1]+=qtde
                cadastro=[]
                atualizei=True
                break
        if not atualizei:
            qtde = int (input ('Digite a quantidade de produtos deste item: '))
            cadastro.append(qtde)
            cadastro_total.append (cadastro)
            cadastro=[]
        naosair = input ('Deseja continuar <S/N>: ')

def item2():
    print ('+---------------------------------------------------------------------+')
    print ('|                         Visualizando Estoque                        |')
    print ('|---------------------------------------------------------------------|')
    print ('|     Nome do Produto','                        ','Quantidade','            |')
    print ('+---------------------------------------------------------------------+')
    for item in cadastro_total:
        print ('|          ',item[0],'                                    ',item[1],'              |')
        print ('+---------------------------------------------------------------------+')

def item3():
    selecao = 'S'
    while selecao.upper() == 'S':
        excluir = input ('Digite o nome do item que deseja excluir: ')
        atualizei=False
        for item in cadastro_total:
            if item[0] == excluir:
                temcerteza = input ('Tem certeza que deseja excluir este item <S/N>? ')
                if temcerteza.upper() == 'S':
                    print ('+---------------------------------------------------------------------+')
                    print ('|                  Exclusão de itens do Estoque                       |')
                    print ('|---------------------------------------------------------------------|')
                    print ('+---------------------------------------------------------------------+')
                    print ('|                    Item encontrado e removido!                      |')
                    print ('+---------------------------------------------------------------------+')
                    cadastro_total.remove(item)
                    atualizei=True
                    break
        if atualizei==False:
            print ('+---------------------------------------------------------------------+')
            print ('|                  Exclusão de itens do Estoque                       |')
            print ('|---------------------------------------------------------------------|')
            print ('+---------------------------------------------------------------------+')
            print ('|                        Item não encontrado!                         |')
            print ('+---------------------------------------------------------------------+')
        selecao = input ('Deseja excluir outro item <S/N>: ')
        if selecao.upper() == 'N':
            break

def item4():
    selecao = 'S'
    while selecao.upper() == 'S':
        escolha = input ('Digite o nome do item que deseja filtrar: ')
        atualizei=False
        for item in cadastro_total:
            if item[0] == escolha:
                print ('+---------------------------------------------------------------------+')
                print ('|                    Filtro de itens do Estoque                       |')
                print ('|---------------------------------------------------------------------|')
                print ('|---------------------------------------------------------------------|')
                print ('|     Nome do Produto','                        ','Quantidade','            |')
                print ('+---------------------------------------------------------------------+')
                print ('|          ',item[0],'                                   ',item[1],'            |')
                print ('+---------------------------------------------------------------------+')
                atualizei=True
                break
        if atualizei==False:
            print ('+---------------------------------------------------------------------+')
            print ('|                    Filtro de itens do Estoque                       |')
            print ('|---------------------------------------------------------------------|')
            print ('+---------------------------------------------------------------------+')
            print ('|                               Item não encontrado!                  |')
            print ('+---------------------------------------------------------------------+')
        escolha = input ('Deseja selecionar outro item <S/N>: ')
        if escolha.upper() == 'N':
            break

def item5():
    nome_arq = input ('Digite o nome do arquivo que será salvo: ')
    with open(nome_arq+'.csv','w',newline='',encoding='utf-8') as arquivo:
        cadastro_writer = csv.writer(arquivo) 
        for linha in cadastro_total:
            cadastro_writer.writerow(linha)

def item6():
    cadastro_total=[]
    nome_arq = input ('Digite o nome do arquivo que deseja carregar: ')
    with open(nome_arq+'.csv','r',newline='',encoding='utf-8') as arquivo:
        cadastro_reader = csv.reader(arquivo) 
        for linha in cadastro_reader:
            cadastro_total.append(linha)
            
opcao = 1000
cadastro_total=[]
while opcao != 7:
    opcao = tela_principal()
    if opcao == 1:
        item1()
    elif opcao == 2:
        item2()
    elif opcao == 3:
        item3()
    elif opcao == 4:
        item4()
    elif opcao == 5:
        item5()
    elif opcao == 6:
        item6()
    elif opcao > 7:
        print ('+---------------------------------------------------------------------+')
        print ('|                Opção Inexistente! Tente Novamente...                |')
        print ('+---------------------------------------------------------------------+')
print ('+---------------------------------------------------------------------+')
print ('|                      Obrigado e até a próxima!                      |')
print ('+---------------------------------------------------------------------+')
```
